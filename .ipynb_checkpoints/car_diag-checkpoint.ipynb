```python
# Car Diagnostics Audio Classification - Complete Analysis
# Multi-class classification of 10 car diagnostic states

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import librosa
import librosa.display
import warnings
warnings.filterwarnings('ignore')

# Machine Learning libraries
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from xgboost import XGBClassifier

# Deep Learning libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.applications import VGG16, ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# For LSTM hybrid
from tensorflow.keras.layers import LSTM, TimeDistributed

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("Libraries loaded successfully!")
```

```python
# Configuration
DATA_PATH = "./data"  # Adjust this path
SAMPLE_RATE = 22050
DURATION = 3  # seconds (adjust based on your files)
N_MELS = 128
N_MFCC = 40
HOP_LENGTH = 512
N_FFT = 2048

# Create output directory for visualizations
os.makedirs("visualizations", exist_ok=True)

print(f"Configuration set. Data path: {DATA_PATH}")
```

```python
# Step 1: Load and explore the dataset
def load_dataset(data_path):
    """
    Load audio files and create a DataFrame with labels
    """
    data = []
    class_mapping = {}
    class_counter = 0
    
    # Walk through directory structure
    for state in os.listdir(data_path):
        state_path = os.path.join(data_path, state)
        if os.path.isdir(state_path):
            for condition in os.listdir(state_path):
                condition_path = os.path.join(state_path, condition)
                if os.path.isdir(condition_path):
                    # Create full label name
                    full_label = f"{state}_{condition}"
                    class_mapping[full_label] = class_counter
                    class_counter += 1
                    
                    # Collect audio files
                    for file in os.listdir(condition_path):
                        if file.endswith('.wav'):
                            file_path = os.path.join(condition_path, file)
                            data.append({
                                'file_path': file_path,
                                'state': state,
                                'condition': condition,
                                'full_label': full_label,
                                'label_id': class_mapping[full_label]
                            })
    
    df = pd.DataFrame(data)
    return df, class_mapping

# Load the data
df, class_mapping = load_dataset(DATA_PATH)
print(f"Loaded {len(df)} audio files")
print(f"Classes: {len(class_mapping)}")
print("\nClass distribution:")
print(df['full_label'].value_counts())

# Display class mapping
print("\nClass Mapping:")
for label, idx in class_mapping.items():
    print(f"  {idx}: {label}")
```

```python
# Step 2: Visualize sample audio files
def plot_waveform_and_spectrogram(file_path, title, save_path=None):
    """
    Plot waveform and mel spectrogram for a single audio file
    """
    # Load audio
    signal, sr = librosa.load(file_path, sr=SAMPLE_RATE, duration=DURATION)
    
    # Create figure
    fig, axes = plt.subplots(2, 1, figsize=(12, 8))
    
    # Plot waveform
    librosa.display.waveshow(signal, sr=sr, ax=axes[0])
    axes[0].set_title(f'Waveform - {title}', fontsize=12)
    axes[0].set_ylabel('Amplitude')
    axes[0].set_xlabel('Time (s)')
    
    # Compute and plot mel spectrogram
    mel_spec = librosa.feature.melspectrogram(y=signal, sr=sr, n_mels=N_MELS, 
                                              n_fft=N_FFT, hop_length=HOP_LENGTH)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    
    img = librosa.display.specshow(mel_spec_db, sr=sr, hop_length=HOP_LENGTH, 
                                   x_axis='time', y_axis='mel', ax=axes[1])
    axes[1].set_title(f'Mel Spectrogram - {title}', fontsize=12)
    fig.colorbar(img, ax=axes[1], format='%+2.0f dB')
    
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()

# Visualize one example from each class
print("Visualizing sample audio files from each class...")
for label in df['full_label'].unique():
    sample_file = df[df['full_label'] == label]['file_path'].iloc[0]
    plot_waveform_and_spectrogram(
        sample_file, 
        label,
        save_path=f"visualizations/sample_{label.replace('/', '_')}.png"
    )
```

```python
# Step 3: Feature extraction functions
def extract_mel_spectrogram(file_path, target_shape=None):
    """
    Extract mel spectrogram as 2D array
    """
    signal, sr = librosa.load(file_path, sr=SAMPLE_RATE, duration=DURATION)
    mel_spec = librosa.feature.melspectrogram(y=signal, sr=sr, n_mels=N_MELS,
                                              n_fft=N_FFT, hop_length=HOP_LENGTH)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    
    if target_shape:
        # Pad or truncate to target shape
        current_shape = mel_spec_db.shape[1]
        if current_shape < target_shape[1]:
            pad_width = target_shape[1] - current_shape
            mel_spec_db = np.pad(mel_spec_db, ((0, 0), (0, pad_width)), mode='constant')
        elif current_shape > target_shape[1]:
            mel_spec_db = mel_spec_db[:, :target_shape[1]]
    
    return mel_spec_db

def extract_mfcc(file_path, n_mfcc=N_MFCC):
    """
    Extract MFCC features
    """
    signal, sr = librosa.load(file_path, sr=SAMPLE_RATE, duration=DURATION)
    mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=n_mfcc,
                                n_fft=N_FFT, hop_length=HOP_LENGTH)
    # Take mean and standard deviation across time
    mfcc_mean = np.mean(mfcc, axis=1)
    mfcc_std = np.std(mfcc, axis=1)
    return np.concatenate([mfcc_mean, mfcc_std])

# Determine target shape for spectrograms
sample_mel = extract_mel_spectrogram(df['file_path'].iloc[0])
TARGET_SHAPE = (N_MELS, sample_mel.shape[1])
print(f"Target spectrogram shape: {TARGET_SHAPE}")
```

```python
# Step 4: Prepare datasets for all approaches
print("Extracting features for all approaches...")

# Extract features
mel_features = []
mfcc_features = []
labels = []

for idx, row in df.iterrows():
    if idx % 50 == 0:
        print(f"Processing {idx}/{len(df)}")
    
    # Extract mel spectrogram
    mel_spec = extract_mel_spectrogram(row['file_path'], TARGET_SHAPE)
    mel_features.append(mel_spec)
    
    # Extract MFCC
    mfcc_feat = extract_mfcc(row['file_path'])
    mfcc_features.append(mfcc_feat)
    
    labels.append(row['label_id'])

# Convert to numpy arrays
mel_features = np.array(mel_features)
mfcc_features = np.array(mfcc_features)
labels = np.array(labels)

# Add channel dimension for CNN (N, height, width, 1)
mel_features_cnn = mel_features.reshape(mel_features.shape[0], mel_features.shape[1], 
                                         mel_features.shape[2], 1)

# Normalize MFCC features
scaler = StandardScaler()
mfcc_features_scaled = scaler.fit_transform(mfcc_features)

print(f"Mel features shape: {mel_features.shape}")
print(f"MFCC features shape: {mfcc_features.shape}")
print(f"Labels shape: {labels.shape}")

# Split data
X_train_mel, X_test_mel, y_train, y_test = train_test_split(
    mel_features_cnn, labels, test_size=0.2, random_state=42, stratify=labels
)

X_train_mfcc, X_test_mfcc, _, _ = train_test_split(
    mfcc_features_scaled, labels, test_size=0.2, random_state=42, stratify=labels
)

# One-hot encode labels for deep learning
y_train_cat = to_categorical(y_train, num_classes=len(class_mapping))
y_test_cat = to_categorical(y_test, num_classes=len(class_mapping))

print(f"Training samples: {len(X_train_mel)}")
print(f"Test samples: {len(X_test_mel)}")
```

```python
# Step 5: Approach 1 - CNN on Mel Spectrograms
def create_cnn_model(input_shape, num_classes):
    """
    Create CNN model for mel spectrogram classification
    """
    model = models.Sequential([
        layers.Input(shape=input_shape),
        
        # First convolutional block
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Second convolutional block
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Third convolutional block
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Fourth convolutional block
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.5),
        
        # Dense layers
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model

# Create and compile model
cnn_model = create_cnn_model((TARGET_SHAPE[0], TARGET_SHAPE[1], 1), len(class_mapping))
cnn_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

cnn_model.summary()

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
]

# Train model
print("Training CNN model...")
history_cnn = cnn_model.fit(
    X_train_mel, y_train_cat,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

# Evaluate
y_pred_cnn = cnn_model.predict(X_test_mel)
y_pred_cnn_classes = np.argmax(y_pred_cnn, axis=1)
cnn_accuracy = accuracy_score(y_test, y_pred_cnn_classes)
cnn_f1 = f1_score(y_test, y_pred_cnn_classes, average='weighted')

print(f"\nCNN Model Results:")
print(f"Accuracy: {cnn_accuracy:.4f}")
print(f"F1-Score: {cnn_f1:.4f}")
```

```python
# Visualize CNN training history
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history_cnn.history['accuracy'], label='Train Accuracy', marker='o')
axes[0].plot(history_cnn.history['val_accuracy'], label='Validation Accuracy', marker='o')
axes[0].set_title('CNN Model Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history_cnn.history['loss'], label='Train Loss', marker='o')
axes[1].plot(history_cnn.history['val_loss'], label='Validation Loss', marker='o')
axes[1].set_title('CNN Model Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig('visualizations/cnn_training_history.png', dpi=150, bbox_inches='tight')
plt.show()

# Confusion matrix for CNN
cm_cnn = confusion_matrix(y_test, y_pred_cnn_classes)
plt.figure(figsize=(12, 10))
sns.heatmap(cm_cnn, annot=True, fmt='d', cmap='Blues',
            xticklabels=list(class_mapping.keys()),
            yticklabels=list(class_mapping.keys()))
plt.title('Confusion Matrix - CNN Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('visualizations/cnn_confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Step 6: Approach 2 - CNN + LSTM Hybrid (FULLY CORRECTED)

def create_sequences_from_spectrogram(mel_spectrograms, freq_dim, time_segments=8):
    """
    Split each spectrogram into equal-sized time segments to create sequences
    """
    n_samples = mel_spectrograms.shape[0]
    time_steps = mel_spectrograms.shape[2]
    
    # Calculate exact segment size (use floor division to ensure equal sizes)
    segment_size = time_steps // time_segments
    
    # If not perfectly divisible, we'll truncate to fit
    if segment_size * time_segments > time_steps:
        segment_size = time_steps // time_segments
    
    sequences = []
    
    for i in range(n_samples):
        spec = mel_spectrograms[i, :, :, 0]  # Remove channel dim: shape (freq_dim, time_steps)
        
        # Truncate to fit exactly into segments
        total_needed = segment_size * time_segments
        if time_steps > total_needed:
            spec = spec[:, :total_needed]
        
        # Split into time segments of equal size
        segments = []
        for t in range(time_segments):
            start = t * segment_size
            end = start + segment_size
            segment = spec[:, start:end]
            segments.append(segment)
        
        # Stack segments for this sample
        sequence = np.stack(segments, axis=0)  # Shape: (time_segments, freq_dim, segment_size)
        sequences.append(sequence)
    
    # Stack all samples
    return np.array(sequences)

# Simplified CNN-LSTM model that's more robust
def create_simplified_cnn_lstm(input_shape, num_classes):
    """
    Simpler CNN-LSTM hybrid model
    """
    model = models.Sequential([
        layers.Input(shape=input_shape),
        
        # CNN layers to extract spatial features from each time segment
        layers.TimeDistributed(layers.Conv2D(32, (3, 3), activation='relu', padding='same')),
        layers.TimeDistributed(layers.BatchNormalization()),
        layers.TimeDistributed(layers.MaxPooling2D((2, 2))),
        layers.TimeDistributed(layers.Dropout(0.25)),
        
        layers.TimeDistributed(layers.Conv2D(64, (3, 3), activation='relu', padding='same')),
        layers.TimeDistributed(layers.BatchNormalization()),
        layers.TimeDistributed(layers.GlobalAveragePooling2D()),
        layers.TimeDistributed(layers.Dropout(0.25)),
        
        # LSTM to process temporal sequence
        layers.LSTM(128, return_sequences=True),
        layers.LSTM(64),
        layers.Dropout(0.5),
        
        # Dense layers for classification
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model

# Prepare data with the corrected function
print("Preparing data for CNN-LSTM hybrid model...")
print(f"Original training data shape: {X_train_mel.shape}")

# Parameters for sequential approach
FREQ_DIM = TARGET_SHAPE[0]  # 128
TIME_SEGMENTS = 8  # Number of time segments

# Create sequential data
print("Creating sequential data for CNN-LSTM...")
X_train_seq = create_sequences_from_spectrogram(X_train_mel, FREQ_DIM, TIME_SEGMENTS)
X_test_seq = create_sequences_from_spectrogram(X_test_mel, FREQ_DIM, TIME_SEGMENTS)

# Add channel dimension back (for CNN)
X_train_seq = X_train_seq.reshape(X_train_seq.shape[0], TIME_SEGMENTS, FREQ_DIM, -1, 1)
X_test_seq = X_test_seq.reshape(X_test_seq.shape[0], TIME_SEGMENTS, FREQ_DIM, -1, 1)

print(f"Sequences training shape: {X_train_seq.shape}")
print(f"Sequences test shape: {X_test_seq.shape}")

# Verify the data is valid
assert not np.any(np.isnan(X_train_seq)), "NaN values found in training data"
assert not np.any(np.isinf(X_train_seq)), "Inf values found in training data"
print("Data validation passed!")

# Create and compile model
cnn_lstm_model = create_simplified_cnn_lstm(
    (TIME_SEGMENTS, FREQ_DIM, X_train_seq.shape[3], 1), 
    len(class_mapping)
)

cnn_lstm_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nCNN-LSTM model summary:")
cnn_lstm_model.summary()

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
]

# Train model
print("\nTraining CNN-LSTM hybrid model...")
print(f"Training samples: {X_train_seq.shape[0]}")
print(f"Validation split: 20%")

history_cnn_lstm = cnn_lstm_model.fit(
    X_train_seq, y_train_cat,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    callbacks=callbacks,
    verbose=1
)

# Evaluate
print("\nEvaluating CNN-LSTM model...")
y_pred_cnn_lstm = cnn_lstm_model.predict(X_test_seq)
y_pred_cnn_lstm_classes = np.argmax(y_pred_cnn_lstm, axis=1)
cnn_lstm_accuracy = accuracy_score(y_test, y_pred_cnn_lstm_classes)
cnn_lstm_f1 = f1_score(y_test, y_pred_cnn_lstm_classes, average='weighted')

print(f"\n{'='*50}")
print(f"CNN-LSTM Model Results:")
print(f"{'='*50}")
print(f"Accuracy: {cnn_lstm_accuracy:.4f} ({cnn_lstm_accuracy*100:.2f}%)")
print(f"F1-Score: {cnn_lstm_f1:.4f}")
print(f"{'='*50}")

# Visualize training history
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history_cnn_lstm.history['accuracy'], label='Train Accuracy', marker='o', linewidth=2)
axes[0].plot(history_cnn_lstm.history['val_accuracy'], label='Validation Accuracy', marker='s', linewidth=2)
axes[0].set_title('CNN-LSTM Model Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Accuracy', fontsize=12)
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

axes[1].plot(history_cnn_lstm.history['loss'], label='Train Loss', marker='o', linewidth=2)
axes[1].plot(history_cnn_lstm.history['val_loss'], label='Validation Loss', marker='s', linewidth=2)
axes[1].set_title('CNN-LSTM Model Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Loss', fontsize=12)
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('visualizations/cnn_lstm_training_history.png', dpi=150, bbox_inches='tight')
plt.show()

# Confusion matrix for CNN-LSTM
cm_cnn_lstm = confusion_matrix(y_test, y_pred_cnn_lstm_classes)
plt.figure(figsize=(12, 10))
sns.heatmap(cm_cnn_lstm, annot=True, fmt='d', cmap='Greens', 
            xticklabels=list(class_mapping.keys()),
            yticklabels=list(class_mapping.keys()),
            square=True)
plt.title('Confusion Matrix - CNN-LSTM Model', fontsize=14, fontweight='bold')
plt.xlabel('Predicted', fontsize=12)
plt.ylabel('Actual', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('visualizations/cnn_lstm_confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

# Print per-class metrics for CNN-LSTM
print("\nDetailed Classification Report for CNN-LSTM:")
print(classification_report(y_test, y_pred_cnn_lstm_classes, 
                           target_names=list(class_mapping.keys())))
```

```python
# Step 7: Approach 3 - MFCC + XGBoost/Random Forest
print("Training Traditional ML models on MFCC features...")

# Random Forest
rf_model = RandomForestClassifier(n_estimators=200, max_depth=20, 
                                  random_state=42, n_jobs=-1)
rf_model.fit(X_train_mfcc, y_train)
y_pred_rf = rf_model.predict(X_test_mfcc)
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf, average='weighted')

print(f"Random Forest Results:")
print(f"Accuracy: {rf_accuracy:.4f}")
print(f"F1-Score: {rf_f1:.4f}")

# XGBoost
xgb_model = XGBClassifier(n_estimators=200, max_depth=10, 
                          learning_rate=0.1, random_state=42)
xgb_model.fit(X_train_mfcc, y_train)
y_pred_xgb = xgb_model.predict(X_test_mfcc)
xgb_accuracy = accuracy_score(y_test, y_pred_xgb)
xgb_f1 = f1_score(y_test, y_pred_xgb, average='weighted')

print(f"\nXGBoost Results:")
print(f"Accuracy: {xgb_accuracy:.4f}")
print(f"F1-Score: {xgb_f1:.4f}")

# Cross-validation for XGBoost
cv_scores = cross_val_score(xgb_model, X_train_mfcc, y_train, 
                           cv=StratifiedKFold(5, shuffle=True, random_state=42))
print(f"\nXGBoost 5-fold CV accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
```

```python
# Step 8: Approach 4 - Transfer Learning (using VGGish-inspired architecture)
def create_transfer_learning_model(input_shape, num_classes):
    """
    Create a model using pretrained weights (simulating VGGish)
    """
    # Use a pretrained model as base (we'll use VGG16 as feature extractor)
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    base_model.trainable = False
    
    # Resize input to match VGG16 expectations
    model = models.Sequential([
        layers.Input(shape=input_shape),
        
        # Resize to 224x224x3
        layers.Lambda(lambda x: tf.image.resize(x, (224, 224))),
        layers.Lambda(lambda x: tf.repeat(x, 3, axis=-1)),  # Convert grayscale to RGB
        
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.5),
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model

# Prepare data for transfer learning (need to resize)
X_train_transfer = X_train_mel.reshape(X_train_mel.shape[0], X_train_mel.shape[1], 
                                        X_train_mel.shape[2], 1)
X_test_transfer = X_test_mel.reshape(X_test_mel.shape[0], X_test_mel.shape[1], 
                                     X_test_mel.shape[2], 1)

# Create and compile model
transfer_model = create_transfer_learning_model((TARGET_SHAPE[0], TARGET_SHAPE[1], 1), 
                                                 len(class_mapping))
transfer_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Training Transfer Learning model...")
history_transfer = transfer_model.fit(
    X_train_transfer, y_train_cat,
    validation_split=0.2,
    epochs=30,
    batch_size=16,
    callbacks=callbacks,
    verbose=1
)

# Fine-tuning: unfreeze some layers of base model
transfer_model.layers[3].trainable = True
transfer_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.00001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nFine-tuning Transfer Learning model...")
history_finetune = transfer_model.fit(
    X_train_transfer, y_train_cat,
    validation_split=0.2,
    epochs=20,
    batch_size=16,
    callbacks=callbacks,
    verbose=1
)

# Evaluate
y_pred_transfer = transfer_model.predict(X_test_transfer)
y_pred_transfer_classes = np.argmax(y_pred_transfer, axis=1)
transfer_accuracy = accuracy_score(y_test, y_pred_transfer_classes)
transfer_f1 = f1_score(y_test, y_pred_transfer_classes, average='weighted')

print(f"\nTransfer Learning Results:")
print(f"Accuracy: {transfer_accuracy:.4f}")
print(f"F1-Score: {transfer_f1:.4f}")
```

```python
# Step 9: Hierarchical classification approach (COMPLETELY CORRECTED)

print("Hierarchical Classification Approach...")

# Create mapping dictionaries
state_names = {0: 'braking_state', 1: 'idle_state', 2: 'startup_state'}
state_to_id = {'braking_state': 0, 'idle_state': 1, 'startup_state': 2}

# First level: classify the state (braking_state, idle_state, startup_state)
state_labels = df['state'].map(state_to_id).values

# Split data for hierarchical approach
X_train_hier, X_test_hier, y_train_state, y_test_state = train_test_split(
    mfcc_features_scaled, state_labels, test_size=0.2, random_state=42, stratify=state_labels
)

# Train state classifier
state_classifier = XGBClassifier(n_estimators=100, max_depth=5, random_state=42, eval_metric='mlogloss')
state_classifier.fit(X_train_hier, y_train_state)
y_pred_state = state_classifier.predict(X_test_hier)
state_accuracy = accuracy_score(y_test_state, y_pred_state)

print(f"State classification accuracy: {state_accuracy:.4f}")

# Create inverse mapping for class names
inv_class_map = {v: k for k, v in class_mapping.items()}

# For each state, train a condition classifier (with remapped labels)
condition_classifiers = {}
label_mappings = {}  # Store mapping from original label to consecutive labels for each state

for state in df['state'].unique():
    print(f"\nTraining classifier for {state}...")
    
    # Get data for this state
    state_data = df[df['state'] == state]
    
    if len(state_data) > 0:
        # Get indices for this state's data
        state_indices = state_data.index
        
        # Extract features for this state
        state_mfcc = mfcc_features_scaled[state_indices]
        
        # Get original labels for this state
        original_labels = state_data['label_id'].values
        
        # Remap labels to consecutive integers starting from 0
        unique_labels = np.unique(original_labels)
        label_mapping = {old_label: new_label for new_label, old_label in enumerate(unique_labels)}
        remapped_labels = np.array([label_mapping[label] for label in original_labels])
        
        # Store the mapping for later use
        label_mappings[state] = label_mapping
        
        # Also store the inverse mapping to get original class names
        inv_label_mapping = {new_label: old_label for old_label, new_label in label_mapping.items()}
        
        print(f"  Original labels: {unique_labels}")
        print(f"  Remapped to: {list(label_mapping.keys())} -> {list(label_mapping.values())}")
        print(f"  Number of classes: {len(unique_labels)}")
        print(f"  Samples: {len(state_mfcc)}")
        
        # Train classifier for this state's conditions
        cond_clf = XGBClassifier(n_estimators=100, max_depth=5, random_state=42, eval_metric='mlogloss')
        cond_clf.fit(state_mfcc, remapped_labels)
        
        # Store both classifier and mapping
        condition_classifiers[state] = {
            'model': cond_clf,
            'label_mapping': label_mapping,
            'inv_mapping': inv_label_mapping
        }

print("\n✓ Hierarchical classifiers trained successfully!")

# Test the hierarchical approach on a few samples
print("\n" + "="*50)
print("Testing Hierarchical Classification on sample data:")
print("="*50)

def predict_hierarchical(audio_features, return_details=False):
    """
    Predict using hierarchical approach
    """
    # Level 1: Predict state
    predicted_state_idx = state_classifier.predict(audio_features.reshape(1, -1))[0]
    predicted_state = state_names[predicted_state_idx]
    
    # Level 2: Predict condition within that state
    state_classifier_data = condition_classifiers[predicted_state]
    cond_model = state_classifier_data['model']
    
    # Get remapped prediction
    remapped_prediction = cond_model.predict(audio_features.reshape(1, -1))[0]
    
    # Convert back to original label ID
    original_label_id = state_classifier_data['inv_mapping'][remapped_prediction]
    
    # Get original class name
    predicted_condition = inv_class_map[original_label_id]
    
    if return_details:
        return predicted_state, predicted_condition, original_label_id
    return predicted_condition

# Test on first 5 test samples
for i in range(min(5, len(X_test_mfcc))):
    # Get true state name
    true_state_idx = y_test_state[i]
    true_state = state_names[true_state_idx]
    
    # Get predictions
    pred_state, pred_condition, pred_id = predict_hierarchical(X_test_mfcc[i], return_details=True)
    
    # Get true condition name
    true_condition = inv_class_map[y_test[i]]
    
    print(f"\nSample {i+1}:")
    print(f"  True: {true_state} -> {true_condition}")
    print(f"  Pred: {pred_state} -> {pred_condition}")
    print(f"  {'✓ Correct' if pred_condition == true_condition else '✗ Incorrect'}")

# Evaluate hierarchical approach on entire test set
print("\n" + "="*50)
print("Evaluating Hierarchical Classification on Full Test Set:")
print("="*50)

hierarchical_predictions = []
hierarchical_details = []

for i in range(len(X_test_mfcc)):
    pred_state, pred_condition, pred_id = predict_hierarchical(X_test_mfcc[i], return_details=True)
    hierarchical_predictions.append(pred_id)
    hierarchical_details.append((pred_state, pred_condition))

hierarchical_accuracy = accuracy_score(y_test, hierarchical_predictions)
hierarchical_f1 = f1_score(y_test, hierarchical_predictions, average='weighted')

print(f"\nHierarchical Classification Results:")
print(f"  Accuracy: {hierarchical_accuracy:.4f} ({hierarchical_accuracy*100:.2f}%)")
print(f"  F1-Score: {hierarchical_f1:.4f}")

# Compare with flat classification (using XGBoost from earlier)
print(f"\nComparison with Flat XGBoost:")
print(f"  Flat XGBoost Accuracy: {xgb_accuracy:.4f}")
print(f"  Hierarchical Accuracy: {hierarchical_accuracy:.4f}")
print(f"  Difference: {hierarchical_accuracy - xgb_accuracy:+.4f}")

# Per-state accuracy analysis
print("\n" + "="*50)
print("Per-State Performance Analysis:")
print("="*50)

for state_name, state_id in state_to_id.items():
    # Get indices for this state in test set
    state_indices = [i for i in range(len(y_test)) if state_names[y_test_state[i]] == state_name]
    
    if state_indices:
        state_true = [y_test[i] for i in state_indices]
        state_pred = [hierarchical_predictions[i] for i in state_indices]
        state_acc = accuracy_score(state_true, state_pred)
        
        # Get condition names for this state
        conditions = df[df['state'] == state_name]['condition'].unique()
        
        print(f"\n{state_name}:")
        print(f"  Accuracy: {state_acc:.4f} ({state_acc*100:.2f}%)")
        print(f"  Samples: {len(state_indices)}")
        print(f"  Conditions: {', '.join(conditions)}")

# Confusion matrix for hierarchical approach
plt.figure(figsize=(12, 10))
cm_hierarchical = confusion_matrix(y_test, hierarchical_predictions)
sns.heatmap(cm_hierarchical, annot=True, fmt='d', cmap='Purples',
            xticklabels=list(class_mapping.keys()),
            yticklabels=list(class_mapping.keys()))
plt.title('Confusion Matrix - Hierarchical Classification', fontsize=14, fontweight='bold')
plt.xlabel('Predicted', fontsize=12)
plt.ylabel('Actual', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('visualizations/hierarchical_confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

# Print detailed classification report for hierarchical approach
print("\nDetailed Classification Report for Hierarchical Approach:")
print(classification_report(y_test, hierarchical_predictions, 
                           target_names=list(class_mapping.keys())))

# Analyze error propagation
print("\n" + "="*50)
print("Error Propagation Analysis:")
print("="*50)

correct_state = 0
wrong_state_correct_condition = 0
wrong_state_wrong_condition = 0

for i in range(len(X_test_mfcc)):
    true_state = state_names[y_test_state[i]]
    pred_state, pred_condition, _ = predict_hierarchical(X_test_mfcc[i], return_details=True)
    
    if pred_state == true_state:
        correct_state += 1
    else:
        # Check if condition would have been correct if state was correct
        # For this analysis, we check if the condition prediction matches
        # the true condition (regardless of state)
        true_condition = inv_class_map[y_test[i]]
        if pred_condition == true_condition:
            wrong_state_correct_condition += 1
        else:
            wrong_state_wrong_condition += 1

total = len(X_test_mfcc)
print(f"\nState Classification (first level):")
print(f"  Correct state: {correct_state}/{total} ({correct_state/total*100:.2f}%)")
print(f"  Wrong state: {total - correct_state}/{total} ({(total-correct_state)/total*100:.2f}%)")

if total - correct_state > 0:
    print(f"\nAmong wrong state predictions ({total-correct_state} samples):")
    print(f"  Condition would have been correct: {wrong_state_correct_condition}/{total-correct_state} ({wrong_state_correct_condition/(total-correct_state)*100:.2f}% of errors)")
    print(f"  Both state and condition wrong: {wrong_state_wrong_condition}/{total-correct_state} ({wrong_state_wrong_condition/(total-correct_state)*100:.2f}% of errors)")

print("\n💡 Insight: Error propagation occurs when state is misclassified,")
print("   but note that conditions are still predicted within the wrong state context.")

# Add hierarchical predictions to results for final comparison
print("\n✓ Hierarchical classification complete!")
```

```python
# Step 12: Model saving and inference example
def save_best_model(results_df, models_dict):
    """
    Save the best performing model
    """
    best_model_name = results_df.iloc[0]['Model']
    print(f"\nSaving best model: {best_model_name}")
    
    # Map model name to actual model object
    model_map = {
        'CNN (Mel Spectrogram)': ('cnn_model', cnn_model),
        'CNN-LSTM Hybrid': ('cnn_lstm_model', cnn_lstm_model),
        'Random Forest (MFCC)': ('rf_model', rf_model),
        'XGBoost (MFCC)': ('xgb_model', xgb_model),
        'Transfer Learning (VGG16)': ('transfer_model', transfer_model)
    }
    
    if best_model_name in model_map:
        model_name, model_obj = model_map[best_model_name]
        
        # Save model
        if best_model_name in ['CNN (Mel Spectrogram)', 'CNN-LSTM Hybrid', 'Transfer Learning (VGG16)']:
            model_obj.save(f'best_model_{model_name}.h5')
            print(f"Saved deep learning model to best_model_{model_name}.h5")
        else:
            import joblib
            joblib.dump(model_obj, f'best_model_{model_name}.pkl')
            print(f"Saved traditional ML model to best_model_{model_name}.pkl")
        
        # Also save the scaler for MFCC features
        joblib.dump(scaler, 'mfcc_scaler.pkl')
        print("Saved MFCC scaler to mfcc_scaler.pkl")

# Example inference function
def predict_audio_file(file_path, model_type='xgb'):
    """
    Predict the diagnostic state from an audio file
    """
    # Extract features
    if model_type in ['xgb', 'rf']:
        features = extract_mfcc(file_path)
        features_scaled = scaler.transform(features.reshape(1, -1))
        
        if model_type == 'xgb':
            prediction = xgb_model.predict(features_scaled)[0]
        else:
            prediction = rf_model.predict(features_scaled)[0]
    else:
        # For deep learning models
        mel_spec = extract_mel_spectrogram(file_path, TARGET_SHAPE)
        mel_spec = mel_spec.reshape(1, TARGET_SHAPE[0], TARGET_SHAPE[1], 1)
        
        if model_type == 'cnn':
            prediction = np.argmax(cnn_model.predict(mel_spec)[0])
        elif model_type == 'cnn_lstm':
            # Reshape for CNN-LSTM
            time_steps = 8
            freq_dim = TARGET_SHAPE[0]
            time_per_step = TARGET_SHAPE[1] // time_steps
            mel_spec = mel_spec.reshape(1, time_steps, freq_dim, time_per_step, 1)
            prediction = np.argmax(cnn_lstm_model.predict(mel_spec)[0])
        elif model_type == 'transfer':
            prediction = np.argmax(transfer_model.predict(mel_spec)[0])
    
    # Get class name
    inv_map = {v: k for k, v in class_mapping.items()}
    return inv_map[prediction]

print("\nModel saving complete!")
print("\nExample usage:")
print("prediction = predict_audio_file('path/to/audio.wav', model_type='xgb')")
```
